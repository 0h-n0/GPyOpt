We compare the optimization of the acquisition function in BO using 3 different strategies:

- 'brute': the aquisition function is evaluated in a grid of points. The maximum is selected and a local optimizer is run in that point

- 'random': several optimizers are run on ramdomly selected locations. The best one is selected.

- 'density': The acquisition function is trated as a density function from which initial points for local optimizers are drawn. Line integrals for local locations and directions are used for this.
