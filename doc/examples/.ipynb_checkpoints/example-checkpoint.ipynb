Examples of use of the class BayesianOptimization
        - forrester (1d)
        - branin (2d))
        - gSobol (arbitrary dimension)

```python
%pylab inline  
# load packages and set seed
import GPy
import GPyOpt
from numpy.random import seed
seed(123)

#tyt create the object function
f_true = GPyOpt.fmodels.experiments1d.forrester()
f_sim = GPyOpt.fmodels.experiments1d.forrester(sd= .5)
f_true.plot()

# box constrains of the function and horizont 
bounds = [(0,1)]
H = 1

# starts the optimization 
myBopt = GPyOpt.methods.BayesianOptimization(f=f_sim.f,bounds=bounds,acquisition='EI',normalize = False, verbosity=0)
myBopt.start_optimization(max_iter = 30,acqu_optimize_method = 'brute',acqu_optimize_restarts = 10, stop_criteria=0.001)
myBopt.plot_acquisition()
myBopt.plot_convergence()
myBopt.x_opt
```
