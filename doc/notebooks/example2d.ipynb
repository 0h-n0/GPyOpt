Optimization of the branin function with GPyOpt

```python
# load packages and set random seed
%pylab inline  
import GPy
import GPyOpt
from numpy.random import seed
import matplotlib.pyplot as plt
seed(1234)
```

```python
# create the object function
f_true = GPyOpt.fmodels.experiments2d.branin()
f_sim = GPyOpt.fmodels.experiments2d.branin(sd = 0.1)
f_true.plot()
bounds = f_true.bounds
```

```python
# Creates three identical objects that we will later use to compare the optimization strategies 
LCB_brute = GPyOpt.methods.BayesianOptimization(f_sim.f,
                                              bounds=bounds,
                                              acquisition='LCB',
                                              acquisition_par = 2,
                                              normalize = False, 
                                              verbosity=0)
```

```python
# runs the optimization for the three methods
LCB_brute.start_optimization(max_iter = 10, 
                          n_inbatch = 2,
                          acqu_optimize_method = 'brute',
                          acqu_optimize_restarts = 4, 
                          stop_criteria=1e-10)
```

```python
LCB_brute.plot_acquisition()
```

```python
LCB_brute.plot_convergence()
```
