Examples of use of the class BayesianOptimization
        - forrester (1d)
        - branin (2d))
        - gSobol (arbitrary dimension)

```python
%pylab inline  
# load packages and set seed
import GPy
import GPyOpt
from numpy.random import seed
import warnings
warnings.filterwarnings('ignore')
seed(123)

#tyt create the object function
f_true = GPyOpt.fmodels.experiments1d.forrester()
f_sim = GPyOpt.fmodels.experiments1d.forrester(sd= .5)
f_true.plot()

# box constrains of the function and horizont 
bounds = [(0,1)]
```

```python
# starts the optimization 
myBopt = GPyOpt.methods.BayesianOptimization(f=f_sim.f,
                                             bounds=bounds,
                                             acquisition='LCB',
                                             acquisition_par = 2,
                                             normalize = False, 
                                             verbosity=0)

myBopt.start_optimization(max_iter = 20, 
                          n_inbatch = 1,
                          acqu_optimize_method = 'brute', 
                          acqu_optimize_restarts = 30, 
                          stop_criteria=-1)

myBopt.plot_acquisition()
myBopt.plot_convergence()
```

```python
# starts the optimization 
myBopt = GPyOpt.methods.BayesianOptimization(f=f_sim.f,
                                             bounds=bounds,
                                             acquisition='LCB',
                                             acquisition_par=2,
                                             normalize = False, 
                                             verbosity=0)

from IPython.display import clear_output
N_iter = 20

for i in range(N_iter):
    clear_output()
    myBopt.start_optimization(max_iter = 1, 
                              acqu_optimize_method = 'brute', 
                              acqu_optimize_restarts = 30, 
                              stop_criteria=-1)
    myBopt.plot_acquisition('iteration%.03i.png' % (i + 1))
```
