# GPyOpt in less that 3 minutes

### Written by Javier Gonzalez (j.h.gonzalez@sheffield.ac.uk), University of Sheffield.

Load GPyOpt

```python
import GPyOpt
```

Define the function that you want to optimize and the boundaries of the problem

```python
def f(x):
    return (2*x)**2

bounds = [(-2,1)]
```

Create a GPyOpt object with the problem

```python
myProblem = GPyOpt.methods.BayesianOptimization(f,bounds)
```

And optimize the function

```python
myProblem.start_optimization()
```

```python
Show the found minimum
```

```python
myProblem.x_opt
```

and to the predicted value value of $f$ at $x^*$ optimum by

```python
myProblem.fx_opt
```
