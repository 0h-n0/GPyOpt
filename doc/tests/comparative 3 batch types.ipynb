```python
%pylab inline  
# load packages and set seed
from GPyOpt.util.general import reshape
from GPyOpt.util.sampling import samples_multimensional_uniform
import GPy
import GPyOpt
from numpy.random import seed
from scipy.stats import norm
from GPyOpt.util.general import multigrid, samples_multidimensional_uniform, reshape
import scipy
from pylab import savefig
seed(123)
```

```python
# create the object function
f_true = GPyOpt.fmodels.experiments2d.branin()
f_sim = GPyOpt.fmodels.experiments2d.branin(sd= 0.1)

f_true.plot()
bounds = f_true.bounds
```

```python
# create object
acquisition_name = 'LCB'
acquisition_par = 2

myBopt = GPyOpt.methods.BayesianOptimization(f=f_sim.f,
                                             bounds=bounds,
                                             acquisition=acquisition_name,
                                             acquisition_par=acquisition_par)

# parameters of the optimization
max_iter = 15
n_inbatch = 5
acqu_optimize_method = 'random' 
acqu_optimize_restarts = 6

batch_method = 'random'
batch_method = 'predmean'
batch_method = 'adaptive'

myBopt.start_optimization(max_iter = max_iter, 
                          batch_method = batch_method,
                          n_inbatch=n_inbatch, 
                          acqu_optimize_method=acqu_optimize_method,
                          acqu_optimize_restarts=acqu_optimize_restarts,
                          stop_criteria = -1)
```

```python
myBopt.plot_convergence()
myBopt.plot_acquisition()
#savefig('acquisition.jpeg')
```
