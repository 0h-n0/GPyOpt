```python
%pylab inline  
import GPy
import GPyOpt
from numpy.random import seed
from scipy.stats import norm
import scipy
from pylab import savefig
seed(123)
```

```python
# create the object function
f_true = GPyOpt.fmodels.experiments2d.sixhumpcamel()
f_sim = GPyOpt.fmodels.experiments2d.sixhumpcamel(sd= 0.1)
f_true.plot()
bounds = f_true.bounds
```

```python
# create object
acquisition_name = 'LCB'
acquisition_par = 2
max_iter = 10
batch_method = 'adaptive'
n_inbatch = 5
acqu_optimize_method = 'random' 
acqu_optimize_restarts = 6


myBopt = GPyOpt.methods.BayesianOptimization(f=f_sim.f,
                                             bounds=bounds,
                                             acquisition=acquisition_name,
                                             acquisition_par=acquisition_par)

myBopt.start_optimization(max_iter = max_iter, 
                          batch_method = batch_method,
                          n_inbatch=n_inbatch, 
                          acqu_optimize_method=acqu_optimize_method,
                          acqu_optimize_restarts=acqu_optimize_restarts,
                          stop_criteria = -1)
```

```python
myBopt.plot_convergence()
myBopt.plot_acquisition()
#savefig('acquisition.jpeg')
```

```python
# functions for plotting
def namean(dat,axis=0):
    mdat = np.ma.masked_array(dat,np.isnan(dat))
    mm = np.mean(mdat,axis=axis)
    return mm.filled(np.nan) # the desired answer

def nastd(dat,axis=0):
    mdat = np.ma.masked_array(dat,np.isnan(dat))
    mm = np.std(mdat,axis=axis)
    return mm.filled(np.nan) # the desired answer

# plots
plt.errorbar(range(max_iter*n_inbatch),namean(best_sequential[:,-(max_iter*n_inbatch):]), yerr = nastd(best_sequential[:,-(max_iter*n_inbatch):])/np.sqrt(max_iter),label='sequential')
plt.errorbar(range(max_iter*n_inbatch),namean(best_random[:,-(max_iter*n_inbatch):]), yerr = nastd(best_random[:,-(max_iter*n_inbatch):])/np.sqrt(max_iter),label='random')
plt.errorbar(range(max_iter*n_inbatch),namean(best_predmean[:,-(max_iter*n_inbatch):]), yerr = nastd(best_predmean[:,-(max_iter*n_inbatch):])/np.sqrt(max_iter),label='predmean')
plt.errorbar(range(max_iter*n_inbatch),namean(best_adaptive[:,-(max_iter*n_inbatch):]), yerr = nastd(best_adaptive[:,-(max_iter*n_inbatch):])/np.sqrt(max_iter),label='adaptive')
plt.legend()
```
