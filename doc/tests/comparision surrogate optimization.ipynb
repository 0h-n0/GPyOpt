We compare the optimization of the acquisition function in BO using 3 different strategies:

- 'brute': the aquisition function is evaluated in a grid of points. The maximum is selected and a local optimizer is run in that point

- 'random': several optimizers are run on ramdomly selected locations. The best one is selected.

- 'density': The acquisition function is trated as a density function from which initial points for local optimizers are drawn. Line integrals for local locations and directions are used for this.

```python
# load packages and set random seed
%pylab inline  
import GPy
import GPyOpt
from numpy.random import seed
import matplotlib.pyplot as plt
seed(12345)
```

```python
# create the object function
f_true = GPyOpt.fmodels.experiments2d.branin()
f_sim = GPyOpt.fmodels.experiments2d.branin(sd = 0.1)
f_true.plot()
bounds = f_true.bounds
```

```python
# parameters of the simulation
NR = 5               # number of replicates
model_data_init = 3    # number of random points to initualize the models
max_iter = 25          # maximum number of collected samples
restarts = 16           # number of initial points for surrogate optimization
sqrt_restarts = 4      # square value (for )
acquisition_par = 0.01
acquisition_function = 'MPI'

# matrices to save the results
best_brute = np.zeros((NR,model_data_init+max_iter+1))
best_random = np.zeros((NR,model_data_init+max_iter+1))
best_fast_brute = np.zeros((NR,model_data_init+max_iter+1))
best_fast_random = np.zeros((NR,model_data_init+max_iter+1))

best_brute[best_brute==0]= np.nan
best_random[best_random==0] = np.nan
best_fast_brute[best_fast_brute==0] = np.nan
best_fast_random[best_fast_random==0] = np.nan


for k in range(NR):
    # create to BayesOpt objects
    LCB_brute = GPyOpt.methods.BayesianOptimization(f_sim.f, bounds=bounds, acquisition=acquisition_function, acquisition_par = acquisition_par)
    LCB_random = GPyOpt.methods.BayesianOptimization(f_sim.f, X = LCB_brute.X, Y = LCB_brute.Y, bounds=bounds,acquisition=acquisition_function, acquisition_par = acquisition_par)
    LCB_fast_brute = GPyOpt.methods.BayesianOptimization(f_sim.f, X = LCB_brute.X, Y = LCB_brute.Y, bounds=bounds,acquisition=acquisition_function, acquisition_par = acquisition_par)
    LCB_fast_random = GPyOpt.methods.BayesianOptimization(f_sim.f, X = LCB_brute.X, Y = LCB_brute.Y, bounds=bounds,acquisition=acquisition_function, acquisition_par = acquisition_par)

    # run the BayesOpt loop
    LCB_brute.start_optimization(max_iter = max_iter, acqu_optimize_method = 'brute',acqu_optimize_restarts = 15, stop_criteria=-1)
    LCB_random.start_optimization(max_iter = max_iter, acqu_optimize_method = 'random', acqu_optimize_restarts = restarts, stop_criteria=-1)
    LCB_fast_brute.start_optimization(max_iter = max_iter, acqu_optimize_method = 'fast_brute',acqu_optimize_restarts = 15, stop_criteria=-1)
    LCB_fast_random.start_optimization(max_iter = max_iter, acqu_optimize_method = 'fast_random', acqu_optimize_restarts = restarts, stop_criteria=-1)
    
    # save results in a matrix
    best_brute[k,:len(LCB_brute.Y_best)] = LCB_brute.Y_best 
    best_random[k,:len(LCB_random.Y_best)] = LCB_random.Y_best 
    best_fast_brute[k,:len(LCB_fast_brute.Y_best)] = LCB_fast_brute.Y_best 
    best_fast_random[k,:len(LCB_fast_random.Y_best)] = LCB_fast_random.Y_best 

best_brute2 = best_brute
```

```python
def namean(dat,axis=0):
    mdat = np.ma.masked_array(dat,np.isnan(dat))
    mm = np.mean(mdat,axis=axis)
    return mm.filled(np.nan) # the desired answer

def nastd(dat,axis=0):
    mdat = np.ma.masked_array(dat,np.isnan(dat))
    mm = np.std(mdat,axis=axis)
    return mm.filled(np.nan) # the desired answer

```

```python
plt.errorbar(range(max_iter),namean(best_brute[:,-max_iter:]), yerr = nastd(best_brute[:,-max_iter:])/np.sqrt(max_iter),label='brute')
plt.errorbar(range(max_iter),namean(best_random[:,-max_iter:]), yerr = nastd(best_random[:,-max_iter:])/np.sqrt(max_iter),label='random')
plt.errorbar(range(max_iter),namean(best_fast_brute[:,-max_iter:]), yerr = nastd(best_fast_brute[:,-max_iter:])/np.sqrt(max_iter),label='fast brute')
plt.errorbar(range(max_iter),namean(best_fast_random[:,-max_iter:]), yerr = nastd(best_fast_random[:,-max_iter:])/np.sqrt(max_iter),label='fast random')
plt.legend()
```
