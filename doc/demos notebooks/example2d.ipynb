Optimization of the branin function with GPyOpt

```python
# load packages and set random seed
%pylab inline  
import GPy
import GPyOpt
from numpy.random import seed
import matplotlib.pyplot as plt
seed(1234)
```

```python
# create the object function
f_true = GPyOpt.fmodels.experiments2d.branin()
f_sim = GPyOpt.fmodels.experiments2d.branin(sd = 0.1)
f_true.plot()
bounds = f_true.bounds
```

```python
# Creates three identical objects that we will later use to compare the optimization strategies 
LCB_brute = GPyOpt.methods.BayesianOptimization(f_sim.f,
                                              bounds=bounds,
                                              acquisition='LCB',
                                              acquisition_par = 2,
                                              normalize = False, 
                                              verbosity=0)
```

```python
# runs the optimization for the three methods
LCB_brute.start_optimization(max_iter = 20, 
                          n_inbatch = 1,
                          acqu_optimize_method = 'brute',
                          acqu_optimize_restarts = 4, 
                          stop_criteria=1e-10)
```

```python
LCB_brute.plot_acquisition()
```

```python
LCB_brute.plot_convergence()
```

```python
# starts the optimization 
myBopt = GPyOpt.methods.BayesianOptimization(f=f_sim.f,
                                             bounds=bounds,
                                             acquisition='LCB',
                                             acquisition_par=2,
                                             normalize = False, 
                                             verbosity=0)

from IPython.display import clear_output
N_iter = 30

for i in range(N_iter):
    print i
    clear_output()
    myBopt.start_optimization(max_iter = 1, 
                              acqu_optimize_method = 'brute', 
                              acqu_optimize_restarts = 30, 
                              stop_criteria=-1)
    myBopt.plot_acquisition('iteration%.03i.png' % (i + 1))
```

```python
# Import JSAnimation 
from JSAnimation import IPython_display, HTMLWriter
from matplotlib import animation
import Image
import glob

# Choose location of files to use
filenames=sorted(glob.glob('iteration*.png'))

#Create figure
fig = plt.figure(figsize=(10,5))
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('off')
im = plt.imshow(Image.open(filenames[0]))

# Create initializatio function
def init():
    im.set_data(Image.open(filenames[0]))
    return im,

# Create animation function
def animate(i):
    image=Image.open(filenames[i])
    im.set_data(image)
    return im,

nframes = 20
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=nframes, interval=1000, blit=True)

#set embed_frames=True to embed base64-encoded frames directly in the HTML
anim.save('animation.html', writer=HTMLWriter(embed_frames=True))

# Show animation in ipython notebook
anim
```
